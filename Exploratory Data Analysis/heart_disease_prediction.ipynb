# Heart Disease Prediction Model
## <ins>Exploratory Data Analysis</ins>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Load data from file
data = pd.read_csv('heart.csv')

# Rename feature labels for readability
column_rename_labels = {
    'cp' : 'chest_pain',
    'trestbps' : 'resting_bp',
    'chol' : 'serum_cholestorol',
    'fbs' : 'fasting_blood_sugar',
    'restecg' : 'resting_ecg',
    'thalach' : 'max_heart_rate',
    'exang' : 'excercise_induced_ang',
    'oldpeak' : 'st_depression',
    'slope' : 'slope_peak_st',
    'ca' : 'major_vessels_colored',
    'thal' : 'thalassemia'
 }

data = data.rename(columns=column_rename_labels)

# Summary statistics
data.describe()
```

## <ins>Visualizations</ins>
### Histograms

```python
# Histogram for age
plt.hist(data['age'], edgecolor='black')
plt.title('Patient Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xticks(range(25, 80, 5))
plt.show()

# Histogram for sex
plt.hist(data['sex'], bins=[-0.5, 0.5, 1.5], edgecolor='black')
plt.title('Patient Sex Distribution')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Frequency')
plt.xticks([0, 1])
plt.show()
```

```python
# Histogram for attributes
data.hist(figsize=(20,20))
plt.savefig('Histogram for attributes')
```

## Boxplots

```python
fig, col = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))  # 1 row, 2 columns

# Box plot for cholesterol levels
sns.boxplot(x="target", y='serum_cholestorol', data=data, ax=col[0])  # Use ax to assign plot to first subplot
col[0].set_title('Cholesterol Boxplot')
col[0].set_xlabel('Heart Disease')
col[0].set_ylabel('Cholesterol')
col[0].set_xticks([0, 1])
col[0].set_xticklabels(['No Disease', 'Disease'])

# Box plot for max heart rate
sns.boxplot(x="target", y='max_heart_rate', data=data, ax=col[1])  # Use ax to assign plot to second subplot
col[1].set_title('Max Heart Rate Boxplot')
col[1].set_xlabel('Heart Disease')
col[1].set_ylabel('Heart Rate')
col[1].set_xticks([0, 1])
col[1].set_xticklabels(['No Disease', 'Disease'])

# Display the plot
plt.tight_layout()  # Adjusts subplot params so that subplots are nicely fit in the figure area.
plt.show()

# Note: Can add more plots if you guys prefered!
```

## Correlation Matrix

```python
# Correlation Matrix for all attributes
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(),linewidth=.01,annot=True,cmap="coolwarm")
plt.show()
plt.savefig('correlationfigure')
```
